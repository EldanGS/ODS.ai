### What are the mean and standard deviation of height for female basketball players participated in the 2000 Olympics? Round the answer to the first decimal.

```python
import math
```

```python
basket_female = data[(data['Year'] == 2000) & (data['Sport'] == 'Basketball') & (data['Sex'] == 'F')].drop_duplicates('Name')
```

```python
type(basket_female)
```

```python
# Mean 
mean = basket_female['Height'].mean()
print(round(mean, 1))
```

```python
def my_std(values):
    sum = 0
    for value in values:
        sum += (value - mean) ** 2
    
    return math.sqrt(sum / (len(values) - 1))
```

```python
# Standard deviation
std = my_std(basket_female['Height'])
print(round(std, 1))
```

```python
# Standard deviation by numpy
print(round(basket_female['Height'].std(), 1))
```
